```python
import pandas as pd
```

1) Ogrenci Bilgileri

```python
py_mind = pd.ExcelFile("py_mind.xlsx")
py_sense = pd.ExcelFile("py_sense.xlsx")
py_science = pd.ExcelFile("py_science.xlsx")
py_opinion = pd.ExcelFile("py_opinion.xlsx")
```

```python
dosyalar = [py_mind, py_sense, py_science, py_opinion]
siniflar=[]
isim=[]
dogru=[]
yanlis=[]
bos=[]

for i in dosyalar:                                     # dort farklı excel dosyası için döngü.
    for j in range(len(i.sheet_names)):                # her dosyanın içindeki sayfa sayısı kadar döngü.
        
        data_mind = pd.read_excel(i, i.sheet_names[j]) # her sayfaya tek tek ulaşma.
        
        isim += [data_mind.columns[0]]                 # istenilen bilgiler hangi satir hangi sütunda ise onları alma.
        dogru += [data_mind.iloc[-3,1]]
        yanlis += [data_mind.iloc[-2,1]]
        bos += [data_mind.iloc[-1,1]]
        
        if i == py_mind:
            sinif = "py_mind"
        elif i == py_sense:
            sinif = "py_sense"
        elif i == py_science:
            sinif = "py_science"
        elif i == py_opinion:
            sinif = "py_opinion"
        
        siniflar += [sinif]

all_students = pd.DataFrame({"Sinif ismi": siniflar, "Ogrenci Adi": isim ,"Dogru Sayisi": dogru, 
                             "Yanlis Sayisi": yanlis, "Bos Sayisi": bos})

all_students.index = all_students.index + 1            # indexleri sıfır yerine bir'den başlatma.
all_students
```

4)Her sinifin ayri ayri en basarili ogrencisini bulunuz.

```python
first_student = all_students[all_students["Sinif ismi"]=='py_mind'].sort_values(by="Dogru Sayisi",ascending=False).head(1)
siniflar = ['py_sense', 'py_science', 'py_opinion']
for i in siniflar:
    first_students = all_students[all_students["Sinif ismi"]==i].sort_values(by="Dogru Sayisi",ascending=False).head(1)
    first_student=pd.concat([first_student,first_students], axis = 0)
first_student
```
